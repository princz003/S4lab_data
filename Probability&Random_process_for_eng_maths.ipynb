### The Average (Central Tendency)
Here we learn the difference between Mean (Average) and Median (Middle value).

- The Plot Twist: Kevin (15 marks) is an Outlier. He drags the Mean down.

- The Hero: The Median is resistant to outliers. It ignores extreme values.

````python
print("\n CHAPTER 4: THE AVERAGE TRAP (Mean vs Median)")
print("="*60)

# Calculate Mean and Median for Math
math_mean = loaded_data['Math'].mean()
math_median = loaded_data['Math'].median()
print("="*60)
print(f" Average (Mean) Math Score: {math_mean:.2f}")
print(f" Middle (Median) Math Score: {math_median:.2f}")
print("="*60)
print("\nCRITICAL THINKING:")
print(f"Notice that the Mean ({math_mean:.2f}) is LOWER than the Median ({math_median:.2f}).")
print("Why? Because Kevin's score of 15 pulled the average down!")
print("The Median didn't care about Kevin. It just looked at the middle student.")
````

### Spread of data (Dispersion)
How "spread out" are the marks?

- Range: Gap between best and worst.

- Standard Deviation (Std Dev): Are students consistent, or are marks all over the place?

````python
print("\n CHAPTER 5: WHO IS CONSISTENT? (Dispersion)")
print("-" * 60)

# Calculate Range
math_range = loaded_data['Math'].max() - loaded_data['Math'].min()
print("="*60)
print(f" Math Range: {math_range} (Gap between Lily and Kevin)")
print("="*60)

# Calculate Standard Deviation
# High Std Dev = Wildly different scores (Unstable class)
# Low Std Dev = Everyone scores similarly (Consistent class)
math_std = loaded_data['Math'].std()
print("="*60)
print(f"Math Standard Deviation: {math_std:.2f}")
print("="*60)
# Let's compare subjects
print("\nWhich subject is the most 'stable'?")
print("="*60)
print(loaded_data[['Math', 'Physics', 'English']].std())
print("="*60)
````

### Finding the Groups (Filtering Data)
Now we use statistics to take action. We use quartiles for Top, middle and lower end groups.

The Creamy Layer: Top 25% (Scholarship candidates).

The Attention Group: Bottom 25% (Needs extra classes).

````python
print("\n CHAPTER 7: FINDING THE CHARACTERS")
print("-" * 60)

# We use 'quantile' to find the cut-off points
# 0.75 quantile means "better than 75% of the class"
top_cutoff = loaded_data['Math'].quantile(0.75)
bottom_cutoff = loaded_data['Math'].quantile(0.25)

print(f"To be in the Creamy Layer, you need > {top_cutoff} marks")
print(f"To fall in the Attention Group, you have < {bottom_cutoff} marks")

print("\nTHE CREAMY LAYER (Top Performers):")
creamy_layer = loaded_data[loaded_data['Math'] > top_cutoff]
display(creamy_layer[['Name', 'Math']])

print("\nNEEDS ATTENTION (Struggling Students):")
attention_group = loaded_data[loaded_data['Math'] < bottom_cutoff]
display(attention_group[['Name', 'Math']])
````

###  Gender Analysis

The Principal has asked a question: "Who performed better on average? The Boys or the Girls?" If we just look at the whole class average, we miss this detail. We need to split the data!

- Filtering: We can create separate lists for Boys and Girls.

- Grouping: We use the groupby commandâ€”the Swiss Army Knife of Pandasâ€”to calculate statistics for each group instantly.

````python
print("\n CHAPTER 8: BOYS vs GIRLS (Group Comparison)")
print("-" * 60)

# METHOD 1: The "Manual" Way (Filtering)
# Let's create a dataframe just for Boys
boys_data = loaded_data[loaded_data['Gender'] == 'M']
girls_data = loaded_data[loaded_data['Gender'] == 'F']

print(f"Number of Boys: {boys_data.shape[0]}")
print(f"Number of Girls: {girls_data.shape[0]}")

# Let's check Math averages manually
boy_math_avg = boys_data['Math'].mean()
girl_math_avg = girls_data['Math'].mean()

print(f"\nManual Check - Math Average:")
print(f"Boys: {boy_math_avg:.2f}")
print(f"Girls: {girl_math_avg:.2f}")

# METHOD 2: The "Pro" Way (Groupby)
# Imagine calculating this for 50 columns! Manual is too slow.
# .groupby() does it all at once.
print("\nThe Power of GroupBy (All Subjects at once):")

# numeric_only=True tells pandas to ignore names/IDs and only calculate math
gender_performance = loaded_data.groupby('Gender').mean(numeric_only=True)
display(gender_performance)

print("\nDETECTIVE INSIGHT:")
print("Look at the Math scores above.")
print("The Boys' average is significantly lower.")
print("Is it because ALL boys are bad at math?")
print("NO! Remember 'Kevin' (15 marks) is a Boy.")
print("One outlier can drag down the whole group's reputation!")

# Let's look at Median to be fair (Median ignores Kevin)
print("\nLet's check the Median (Removing the 'Kevin Effect'):")
gender_median = loaded_data.groupby('Gender').median(numeric_only=True)
display(gender_median)
print("Notice how the gap closes when we use Median?")
````

# Measures of Central Tendency with Visualization

---



````python
# ============================================================
# Exercise No: 2
# Measures of Central Tendency with Visualization
# ============================================================

# ------------------------------------------------------------
# GIVEN DATA
data = [12, 15, 18, 20, 22, 15, 18, 25, 30, 18]
print("Given Data:", data)

# ------------------------------------------------------------
# IMPORT REQUIRED LIBRARIES
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# ------------------------------------------------------------
# FUNCTION TO CALCULATE MEAN
def calculate_mean(data):
    """
    Mean = (Sum of all observations) / (Number of observations)
    """
    return np.sum(data) / len(data)

# ------------------------------------------------------------
# FUNCTION TO CALCULATE MEDIAN
def calculate_median(data):
    """
    Median is the middle value of the sorted data
    """
    sorted_data = sorted(data)
    n = len(sorted_data)

    if n % 2 == 0:
        median = (sorted_data[n//2 - 1] + sorted_data[n//2]) / 2
    else:
        median = sorted_data[n//2]

    return median

# ------------------------------------------------------------
# FUNCTION TO CALCULATE MODE
def calculate_mode(data):
    """
    Mode is the value with highest frequency
    """
    frequency = Counter(data)
    mode = frequency.most_common(1)[0][0]
    return mode

# ------------------------------------------------------------
# CALCULATE MEASURES OF CENTRAL TENDENCY
mean_value = calculate_mean(data)
median_value = calculate_median(data)
mode_value = calculate_mode(data)

# ------------------------------------------------------------
# DISPLAY RESULTS
print("\nMeasures of Central Tendency:")
print("Mean   :", mean_value)
print("Median :", median_value)
print("Mode   :", mode_value)

# ------------------------------------------------------------
# VISUALIZATION
# ------------------------------------------------------------

# HISTOGRAM â€“ Data Distribution
plt.figure()
plt.hist(data, bins=6)
plt.xlabel("Data Values")
plt.ylabel("Frequency")
plt.title("Histogram of Given Data")
plt.show()

# BOXPLOT â€“ Spread and Central Value
plt.figure()
plt.boxplot(data, vert=False)
plt.xlabel("Data Values")
plt.title("Boxplot of Given Data")
plt.show()

# ------------------------------------------------------------
# INTERPRETATION (FOR STUDENTS):
# - Histogram shows frequency distribution of data
# - Boxplot shows median, range, and possible outliers
# - Mean is affected by extreme values
# - Median gives the central value
# - Mode shows the most repeated observation

# ------------------------------------------------------------
# CONCLUSION:
# Statistical measures along with visualization provide
# a clear understanding of the data distribution and
# central tendency in engineering applications.
# ============================================================
````

````python
# ============================================================
# Exercise No: 2
# Measures of Central Tendency using Pandas & NumPy
# (From Excel Data â€“ Engineering Mathematics Lab)
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ------------------------------------------------------------
# AIM:
# To calculate Mean, Median, and Mode from data read
# from an Excel file and visualize the data.

# ------------------------------------------------------------
# READ DATA FROM EXCEL FILE
filename = "/content/drive/MyDrive/Marks.xlsx"

# Read Excel file (no header assumed)
df = pd.read_excel(filename, header=None)

# Convert DataFrame to NumPy array
data = df.values.astype(float)

# Transpose data (each row represents one dataset)
D = data.T

print("Original Data:\n", data)
print("\nTransposed Data:\n", D)

# ------------------------------------------------------------
# MEASURES OF CENTRAL TENDENCY
# ------------------------------------------------------------

# 1. MEAN
# Mean = Average value of the data
mean_values = np.mean(D, axis=1)

# 2. MEDIAN
# Median = Middle value after sorting
median_values = np.median(D, axis=1)

# 3. MODE
# Mode = Most frequently occurring value
def calculate_mode(arr):
    values, counts = np.unique(arr, return_counts=True)
    return values[np.argmax(counts)]

mode_values = np.apply_along_axis(calculate_mode, 1, D)

# ------------------------------------------------------------
# DISPLAY RESULTS
print("\nMeasures of Central Tendency (Row-wise after Transpose):")
for i in range(len(mean_values)):
    print(f"\nData Set {i+1}:")
    print("Mean   =", mean_values[i])
    print("Median =", median_values[i])
    print("Mode   =", mode_values[i])

# ------------------------------------------------------------
# VISUALIZATION
# ------------------------------------------------------------

# HISTOGRAM â€“ Distribution of marks
plt.figure()
plt.hist(D.flatten(), bins=10)
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Histogram of Marks Distribution")
plt.show()

# BOXPLOT â€“ Spread and outliers
plt.figure()
plt.boxplot(D.flatten(), vert=False)
plt.xlabel("Marks")
plt.title("Boxplot of Marks")
plt.show()

# ------------------------------------------------------------
# INTERPRETATION (FOR STUDENTS):
# - Histogram shows frequency distribution of marks
# - Boxplot shows median, spread, and possible outliers
# - Mean is affected by extreme values
# - Median is robust to outliers
# - Mode represents the most frequent score

# ------------------------------------------------------------
# CONCLUSION:
# Measures of central tendency along with visualization
# provide a clear understanding of data distribution and
# student performance in engineering statistics.
# ============================================================
````

````python
import math

def get_data():
    n = int(input("Enter number of data values: "))
    data = []

    for i in range(n):
        value = float(input(f"Enter value {i+1}: "))
        data.append(value)

    return data
#---------------------------------------
# or we can give Data as
# def get_data():
#     """
#     Returns the dataset
#     """
#     return [12, 15, 20, 22, 25, 30, 35]

def calculate_mean(data):
    """
    Calculates arithmetic mean
    """
    return sum(data) / len(data)


def calculate_range(data):
    """
    Calculates range
    """
    return max(data) - min(data)


def calculate_std_dev(data):
    """
    Calculates standard deviation using formula
    Ïƒ = sqrt( (1/n) * Î£ (xi - xÌ„)^2 )
    """
    mean = calculate_mean(data)
    n = len(data)

    total = 0
    for x in data:
        total += (x - mean) ** 2

    variance = total / n
    return math.sqrt(variance)


def display_results(data, data_range, std_dev):
    """
    Displays results clearly
    """
    print("-" * 60)
    print("\nEntered Data        :", data)
    print("-" * 60)
    print("Range of Data      :", data_range)
    print("-" * 60)
    print("Standard Deviation :", round(std_dev, 2))
    print("-" * 60)


# -------- Program Execution --------
data = get_data()
data_range = calculate_range(data)
std_dev = calculate_std_dev(data)
display_results(data, data_range, std_dev)
````

# **Theory (Grouped Data)**

When data is given in class intervals with frequencies, individual values are not directly available.
So we use class mid-points.

````python
import math

def get_grouped_data():
    """
    Takes class intervals and frequencies as input
    """
    n = int(input("Enter number of class intervals: "))

    classes = []
    frequencies = []

    for i in range(n):
        print(f"\nClass interval {i+1}")
        lower = float(input("Enter lower limit: "))
        upper = float(input("Enter upper limit: "))
        freq = int(input("Enter frequency: "))

        classes.append((lower, upper))
        frequencies.append(freq)

    return classes, frequencies


def calculate_midpoints(classes):
    midpoints = []
    for c in classes:
        mid = (c[0] + c[1]) / 2
        midpoints.append(mid)
    return midpoints


def calculate_mean(midpoints, frequencies):
    total_fx = 0
    total_f = sum(frequencies)

    for i in range(len(midpoints)):
        total_fx += midpoints[i] * frequencies[i]

    return total_fx / total_f


def calculate_std_dev_grouped(midpoints, frequencies):
    """
    Ïƒ = sqrt( Î£ f(x âˆ’ xÌ„)Â² / Î£ f )
    """
    mean = calculate_mean(midpoints, frequencies)
    total_f = sum(frequencies)

    sum_fxd2 = 0
    for i in range(len(midpoints)):
        sum_fxd2 += frequencies[i] * (midpoints[i] - mean) ** 2

    variance = sum_fxd2 / total_f
    return math.sqrt(variance)


def display_results(classes, frequencies, midpoints, std_dev):
    print("\nClass Intervals :", classes)
    print("Frequencies    :", frequencies)
    print("Midpoints      :", midpoints)
    print("Standard Deviation :", round(std_dev, 2))


# -------- Program Execution --------
classes, frequencies = get_grouped_data()
midpoints = calculate_midpoints(classes)
std_dev = calculate_std_dev_grouped(midpoints, frequencies)
display_results(classes, frequencies, midpoints, std_dev)
````

Application Problem: Sensor Accuracy Analysis in a Manufacturing Plant.

**Problem Statement**

An industrial engineering team is analysing the temperature readings from a sensor
installed in a high-precision manufacturing machine. The sensor records the temperature
every hour. The following temperature readings (in Â°C) were recorded over a 10-hour shift
are 22.4, 22.7, 22.5, 22.9, 22.8, 22.6, 22.7, 22.3, 22.9, 22.6. Calculate the Range and
standard deviation of the temperature readings.

````python
import math

# Data: Rod diameters in mm
rod_diameters = [
    10.01, 10.03, 9.98, 10.05, 10.00,
    10.02, 10.04, 9.99, 10.01, 10.03,
    9.97, 10.06, 10.02, 10.01, 9.98,
    10.00, 10.04, 9.96, 10.02, 10.01
]

# Calculate range
range_rod = max(rod_diameters) - min(rod_diameters)

# Calculate mean
mean_rod = sum(rod_diameters) / len(rod_diameters)

# Calculate standard deviation (population)
sum_squared_diff = sum((x - mean_rod) ** 2 for x in rod_diameters)
std_dev_rod = math.sqrt(sum_squared_diff / len(rod_diameters))

# Display results
print("Rod Diameters (mm) :", rod_diameters)
print("Range (mm)         :", round(range_rod, 3))
print("Standard Deviation :", round(std_dev_rod, 4))
````
