###  Gender Analysis

The Principal has asked a question: "Who performed better on average? The Boys or the Girls?" If we just look at the whole class average, we miss this detail. We need to split the data!

- Filtering: We can create separate lists for Boys and Girls.

- Grouping: We use the groupby command—the Swiss Army Knife of Pandas—to calculate statistics for each group instantly.

````python
print("\n CHAPTER 8: BOYS vs GIRLS (Group Comparison)")
print("-" * 60)

# METHOD 1: The "Manual" Way (Filtering)
# Let's create a dataframe just for Boys
boys_data = loaded_data[loaded_data['Gender'] == 'M']
girls_data = loaded_data[loaded_data['Gender'] == 'F']

print(f"Number of Boys: {boys_data.shape[0]}")
print(f"Number of Girls: {girls_data.shape[0]}")

# Let's check Math averages manually
boy_math_avg = boys_data['Math'].mean()
girl_math_avg = girls_data['Math'].mean()

print(f"\nManual Check - Math Average:")
print(f"Boys: {boy_math_avg:.2f}")
print(f"Girls: {girl_math_avg:.2f}")

# METHOD 2: The "Pro" Way (Groupby)
# Imagine calculating this for 50 columns! Manual is too slow.
# .groupby() does it all at once.
print("\nThe Power of GroupBy (All Subjects at once):")

# numeric_only=True tells pandas to ignore names/IDs and only calculate math
gender_performance = loaded_data.groupby('Gender').mean(numeric_only=True)
display(gender_performance)

print("\nDETECTIVE INSIGHT:")
print("Look at the Math scores above.")
print("The Boys' average is significantly lower.")
print("Is it because ALL boys are bad at math?")
print("NO! Remember 'Kevin' (15 marks) is a Boy.")
print("One outlier can drag down the whole group's reputation!")

# Let's look at Median to be fair (Median ignores Kevin)
print("\nLet's check the Median (Removing the 'Kevin Effect'):")
gender_median = loaded_data.groupby('Gender').median(numeric_only=True)
display(gender_median)
print("Notice how the gap closes when we use Median?")
````

# Measures of Central Tendency with Visualization

---



## Aim  
To compute the **mean**, **median**, and **mode** of a given dataset using Python, and to visualize the data distribution using **histogram** and **boxplot** to understand central tendency and spread.

## Objective  
1. To calculate the three primary measures of central tendency:  
   - Mean (arithmetic average)  
   - Median (middle value)  
   - Mode (most frequent value)  
2. To implement these calculations manually (without built-in functions where possible) for better conceptual understanding.  
3. To visualize the dataset using histogram (to show frequency distribution) and boxplot (to show median, quartiles, range, and outliers).  
4. To interpret how outliers affect mean vs. median, and how mode reflects repetition in data.  
5. To relate these measures to real-world engineering applications (e.g., quality control, sensor data analysis, reliability studies).

## Theory (Brief)  
- **Mean**: Sum of all observations divided by the number of observations. Sensitive to extreme values (outliers).  
  Formula: μ = (Σxᵢ) / n  
- **Median**: The middle value when data is arranged in ascending order. Robust to outliers.  
  - If n is odd → middle term  
  - If n is even → average of two middle terms  
- **Mode**: The value that appears most frequently in the dataset. A dataset can be unimodal, bimodal, or multimodal.  
- **Histogram**: Graphical representation of frequency distribution.  
- **Boxplot**: Summarizes data using five-number summary (minimum, Q1, median, Q3, maximum) and shows outliers.

## Algorithm  
1. Import necessary libraries:  
   - `numpy` for numerical operations  
   - `matplotlib.pyplot` for visualization  
   - `collections.Counter` for frequency count (mode)  
2. Define the given dataset as a list.  
3. Compute **Mean**:  
   - Sum all elements using `np.sum()`  
   - Divide by the length of the list (`len(data)`)  
4. Compute **Median**:  
   - Sort the data in ascending order  
   - If length is odd → take middle element  
   - If length is even → take average of two middle elements  
5. Compute **Mode**:  
   - Use `Counter` to count frequency of each value  
   - Find the value with highest frequency using `most_common(1)`  
6. Display the calculated mean, median, and mode with proper formatting.  
7. Visualize the data:  
   - Plot a **histogram** to show frequency distribution (choose appropriate bins)  
   - Plot a **boxplot** to show central value, spread, and possible outliers  
8. Interpret the results:  
   - Compare mean vs. median (to check effect of outliers)  
   - Note the mode (most frequent value)  
   - Analyze the shape of histogram and boxplot

````python
# ============================================================
# Exercise No: 2
# Measures of Central Tendency with Visualization
# ============================================================

# ------------------------------------------------------------
# GIVEN DATA
data = [12, 15, 18, 20, 22, 15, 18, 25, 30, 18]
print("Given Data:", data)

# ------------------------------------------------------------
# IMPORT REQUIRED LIBRARIES
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# ------------------------------------------------------------
# FUNCTION TO CALCULATE MEAN
def calculate_mean(data):
    """
    Mean = (Sum of all observations) / (Number of observations)
    """
    return np.sum(data) / len(data)

# ------------------------------------------------------------
# FUNCTION TO CALCULATE MEDIAN
def calculate_median(data):
    """
    Median is the middle value of the sorted data
    """
    sorted_data = sorted(data)
    n = len(sorted_data)

    if n % 2 == 0:
        median = (sorted_data[n//2 - 1] + sorted_data[n//2]) / 2
    else:
        median = sorted_data[n//2]

    return median

# ------------------------------------------------------------
# FUNCTION TO CALCULATE MODE
def calculate_mode(data):
    """
    Mode is the value with highest frequency
    """
    frequency = Counter(data)
    mode = frequency.most_common(1)[0][0]
    return mode

# ------------------------------------------------------------
# CALCULATE MEASURES OF CENTRAL TENDENCY
mean_value = calculate_mean(data)
median_value = calculate_median(data)
mode_value = calculate_mode(data)

# ------------------------------------------------------------
# DISPLAY RESULTS
print("\nMeasures of Central Tendency:")
print("Mean   :", mean_value)
print("Median :", median_value)
print("Mode   :", mode_value)

# ------------------------------------------------------------
# VISUALIZATION
# ------------------------------------------------------------

# HISTOGRAM – Data Distribution
plt.figure()
plt.hist(data, bins=6)
plt.xlabel("Data Values")
plt.ylabel("Frequency")
plt.title("Histogram of Given Data")
plt.show()

# BOXPLOT – Spread and Central Value
plt.figure()
plt.boxplot(data, vert=False)
plt.xlabel("Data Values")
plt.title("Boxplot of Given Data")
plt.show()

# ------------------------------------------------------------
# INTERPRETATION (FOR STUDENTS):
# - Histogram shows frequency distribution of data
# - Boxplot shows median, range, and possible outliers
# - Mean is affected by extreme values
# - Median gives the central value
# - Mode shows the most repeated observation

# ------------------------------------------------------------
# CONCLUSION:
# Statistical measures along with visualization provide
# a clear understanding of the data distribution and
# central tendency in engineering applications.
# ============================================================
````

````python
# ============================================================
# Exercise No: 2
# Measures of Central Tendency using Pandas & NumPy
# (From Excel Data – Engineering Mathematics Lab)
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ------------------------------------------------------------
# AIM:
# To calculate Mean, Median, and Mode from data read
# from an Excel file and visualize the data.

# ------------------------------------------------------------
# READ DATA FROM EXCEL FILE
filename = "/content/drive/MyDrive/Marks.xlsx"

# Read Excel file (no header assumed)
df = pd.read_excel(filename, header=None)

# Convert DataFrame to NumPy array
data = df.values.astype(float)

# Transpose data (each row represents one dataset)
D = data.T

print("Original Data:\n", data)
print("\nTransposed Data:\n", D)

# ------------------------------------------------------------
# MEASURES OF CENTRAL TENDENCY
# ------------------------------------------------------------

# 1. MEAN
# Mean = Average value of the data
mean_values = np.mean(D, axis=1)

# 2. MEDIAN
# Median = Middle value after sorting
median_values = np.median(D, axis=1)

# 3. MODE
# Mode = Most frequently occurring value
def calculate_mode(arr):
    values, counts = np.unique(arr, return_counts=True)
    return values[np.argmax(counts)]

mode_values = np.apply_along_axis(calculate_mode, 1, D)

# ------------------------------------------------------------
# DISPLAY RESULTS
print("\nMeasures of Central Tendency (Row-wise after Transpose):")
for i in range(len(mean_values)):
    print(f"\nData Set {i+1}:")
    print("Mean   =", mean_values[i])
    print("Median =", median_values[i])
    print("Mode   =", mode_values[i])

# ------------------------------------------------------------
# VISUALIZATION
# ------------------------------------------------------------

# HISTOGRAM – Distribution of marks
plt.figure()
plt.hist(D.flatten(), bins=10)
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Histogram of Marks Distribution")
plt.show()

# BOXPLOT – Spread and outliers
plt.figure()
plt.boxplot(D.flatten(), vert=False)
plt.xlabel("Marks")
plt.title("Boxplot of Marks")
plt.show()


````

````python
import math

def get_data():
    n = int(input("Enter number of data values: "))
    data = []

    for i in range(n):
        value = float(input(f"Enter value {i+1}: "))
        data.append(value)

    return data
#---------------------------------------
# or we can give Data as
# def get_data():
#     """
#     Returns the dataset
#     """
#     return [12, 15, 20, 22, 25, 30, 35]

def calculate_mean(data):
    """
    Calculates arithmetic mean
    """
    return sum(data) / len(data)


def calculate_range(data):
    """
    Calculates range
    """
    return max(data) - min(data)


def calculate_std_dev(data):
    """
    Calculates standard deviation using formula
    σ = sqrt( (1/n) * Σ (xi - x̄)^2 )
    """
    mean = calculate_mean(data)
    n = len(data)

    total = 0
    for x in data:
        total += (x - mean) ** 2

    variance = total / n
    return math.sqrt(variance)


def display_results(data, data_range, std_dev):
    """
    Displays results clearly
    """
    print("-" * 60)
    print("\nEntered Data        :", data)
    print("-" * 60)
    print("Range of Data      :", data_range)
    print("-" * 60)
    print("Standard Deviation :", round(std_dev, 2))
    print("-" * 60)


# -------- Program Execution --------
data = get_data()
data_range = calculate_range(data)
std_dev = calculate_std_dev(data)
display_results(data, data_range, std_dev)
````

# **Theory (Grouped Data)**

When data is given in class intervals with frequencies, individual values are not directly available.
So we use class mid-points.

````python
import math

def get_grouped_data():
    """
    Takes class intervals and frequencies as input
    """
    n = int(input("Enter number of class intervals: "))

    classes = []
    frequencies = []

    for i in range(n):
        print(f"\nClass interval {i+1}")
        lower = float(input("Enter lower limit: "))
        upper = float(input("Enter upper limit: "))
        freq = int(input("Enter frequency: "))

        classes.append((lower, upper))
        frequencies.append(freq)

    return classes, frequencies


def calculate_midpoints(classes):
    midpoints = []
    for c in classes:
        mid = (c[0] + c[1]) / 2
        midpoints.append(mid)
    return midpoints


def calculate_mean(midpoints, frequencies):
    total_fx = 0
    total_f = sum(frequencies)

    for i in range(len(midpoints)):
        total_fx += midpoints[i] * frequencies[i]

    return total_fx / total_f


def calculate_std_dev_grouped(midpoints, frequencies):
    """
    σ = sqrt( Σ f(x − x̄)² / Σ f )
    """
    mean = calculate_mean(midpoints, frequencies)
    total_f = sum(frequencies)

    sum_fxd2 = 0
    for i in range(len(midpoints)):
        sum_fxd2 += frequencies[i] * (midpoints[i] - mean) ** 2

    variance = sum_fxd2 / total_f
    return math.sqrt(variance)


def display_results(classes, frequencies, midpoints, std_dev):
    print("\nClass Intervals :", classes)
    print("Frequencies    :", frequencies)
    print("Midpoints      :", midpoints)
    print("Standard Deviation :", round(std_dev, 2))


# -------- Program Execution --------
classes, frequencies = get_grouped_data()
midpoints = calculate_midpoints(classes)
std_dev = calculate_std_dev_grouped(midpoints, frequencies)
display_results(classes, frequencies, midpoints, std_dev)
````

# **Result and Discussion:**

The measures of dispersion provide insights into the spread and variability within the
given dataset. A higher standard deviation suggests a wider spread of data points, while a
lower standard deviation indicates that data points are clustered more closely around the
mean. In this case, a standard deviation of 4.96 suggests moderate variability within the
dataset.

# Calculate the probability of successes of a random variable which follows Binomial distribution

---



=============================================================================

**Experiment: Binomial Probability using scipy.stats**
# **Aim:**
  To calculate the probability of obtaining exactly 'x' successes
  in 'n' independent Bernoulli trials using the Binomial distribution.

# **Objective:**

  Given:

- n = number of trials
- p = probability of success in each trial
- x = number of successes we want
  Compute the probability P(X = x) using the binomial probability mass function.

**Formula (for reference):**

  P(X = x) = C(n,x) × p^x × (1-p)^(n-x)
  
  where C(n,x) is the binomial coefficient "n choose x"


# **Algorithm:**
* Step 1: Import the required function (Import binom function) from scipy.stats.
* Step 2: Define the parameters (as given in the case).
* Step 3: Calculate the probability using binom.pmf(), ie, Calculate exact probability P(X = x).
* Step 4: Display the result in a clear, formatted way.
* Step 4: Visualize the complete Probability Mass Function (PMF)

````python
# =============================================================================
# Binomial Probability Distribution
# =============================================================================

# Import required libraries
from scipy.stats import binom
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define parameters
n = 10          # Total number of independent trials
p = 0.5         # Probability of success in each trial (fair coin)
x = 6           # Desired number of successes

# Step 2: Calculate exact probability P(X = x)
probability = binom.pmf(k=x, n=n, p=p)

# Step 3: Display results
print("Binomial Probability Calculation")
print("-" * 50)
print(f"Number of trials (n)          : {n}")
print(f"Probability of success (p)    : {p}")
print(f"Desired number of successes   : {x}")
print("-" * 50)
print(f"Probability of getting exactly {x} successes   : {probability:.6f}")
print(f"Formatted (4 decimal places)                  : {probability:.4f}")
print(f"Scientific notation                           : {probability:.2e}")
print("-" * 50)

# Step 4: Visualize the complete Probability Mass Function (PMF)
# Generate all possible values of x (0 to n)
possible_x = np.arange(0, n + 1)
pmf_values = binom.pmf(possible_x, n, p)

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(possible_x, pmf_values, color='skyblue', edgecolor='black', alpha=0.8)

# Highlight the desired x value
plt.axvline(x=x, color='red', linestyle='--', linewidth=2.5,
            label=f'x = {x} (P = {probability:.4f})')

# Labels and formatting
plt.title(f'Binomial Probability Mass Function\n(n = {n}, p = {p})', fontsize=14, fontweight='bold')
plt.xlabel('Number of Successes (x)', fontsize=12)
plt.ylabel('Probability P(X = x)', fontsize=12)
plt.xticks(possible_x)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(fontsize=11)
plt.tight_layout()

# Show the plot
plt.show()
````

````python
# Alternative 1: Stem + Line plot (very common in probability textbooks)

plt.figure(figsize=(10, 6))
plt.stem(possible_x, pmf_values, linefmt='b-', markerfmt='bo', basefmt=' ',
         label='P(X = x)')
plt.plot(possible_x, pmf_values, 'r--', alpha=0.6, label='Connecting line')

plt.axvline(x=x, color='green', linestyle='-', linewidth=2.5,
            label=f'x = {x} (P = {probability:.4f})')

plt.title(f'Binomial Distribution – Stem & Line Plot\n(n = {n}, p = {p})', fontsize=14)
plt.xlabel('Number of Successes (x)', fontsize=12)
plt.ylabel('Probability P(X = x)', fontsize=12)
plt.xticks(possible_x)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=11)
plt.tight_layout()
plt.show()
````

# **Result and Discussion:**

The result (0.2051) means there's about a 20.51% chance of getting exactly 6 successes (e.g., heads) in 10 trials (e.g., coin flips) when the probability of success in each trial is 0.5 (e.g., a fair coin).

# Calculate the probability of an event of a random variable which follows Poisson distribution.

---



====================================================================

Aim:

To calculate the probability of a specific number of events occurring in a fixed interval of time or space using the Poisson distribution.

Objective:

To determine the probability of observing exactly x events within a given timeframe or space, assuming the events follow a Poisson distribution with an average rate of lmbda.



# **Algorithm:**

Step 1:Import the poisson function from scipy.stats module.

Step 2:Define the parameter.

Step 3:Calculate the probability mass function for the given x and lambda

Step 4: Display the result.

Case:Find the probability of observing exactly 2 events when the average rate is 4 by using poisson distribution.

````python
# =====================================================
# Poisson Distribution with Visualization
# =====================================================

# Import required libraries
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt

# Step 1: Define parameters
lambda_val = 4     # Average rate (λ)
x = 3              # Exact number of events

# Step 2: Calculate probability
probability = poisson.pmf(x, lambda_val)

print(f"Probability of observing exactly {x} events:")
print(f"P(X = {x}) = {probability:.4f}")

# -----------------------------------------------------
# Step 3: PMF Visualization (Probability Mass Function)
# -----------------------------------------------------

# Range of x values
x_values = np.arange(0, 15)

# PMF values
pmf_values = poisson.pmf(x_values, lambda_val)

plt.figure()
plt.bar(x_values, pmf_values)
plt.xlabel("Number of Events (x)")
plt.ylabel("Probability P(X = x)")
plt.title("Poisson Distribution (PMF) with λ = 4")
plt.show()

# -----------------------------------------------------
# Step 4: CDF Visualization (Cumulative Distribution)
# -----------------------------------------------------

# CDF values
cdf_values = poisson.cdf(x_values, lambda_val)

plt.figure()
plt.step(x_values, cdf_values)
plt.xlabel("Number of Events (x)")
plt.ylabel("Cumulative Probability P(X ≤ x)")
plt.title("Poisson Distribution (CDF) with λ = 4")
plt.show()
````

# Find the probability of a continuous random variable which follows normal distribution

---



# **Aim**

To compute probabilities for a normally distributed continuous random variable falling within specified intervals using Python and to apply it to real-world problems.

**Objective**

* To understand and implement the cumulative distribution function (CDF) of the normal distribution.
* To calculate probabilities for intervals, left tail, and right tail events.
* To estimate number of observations in different ranges for large samples.
* To visualize the normal distribution and shaded probability regions.

# **Theory**

If X ~ N(μ, σ²), then the probability

**P(a ≤ X ≤ b) = F(b) – F(a)**

where F(x) = P(X ≤ x) is the cumulative distribution function.
For the standard normal distribution Z ~ N(0,1):

**P(a ≤ Z ≤ b) = Φ(b) – Φ(a)**

where Φ(·) is the standard normal CDF.

Any normal variable can be standardized:

**Z = (X – μ) / σ → Z ~ N(0,1)**

# **Algorithm**

* Import required libraries: scipy.stats.norm, numpy, matplotlib.pyplot
* Define population parameters: mean (μ), standard deviation (σ)
* Define the interval limits (a, b) or tail condition
* Compute CDF values using ***norm.cdf(x, loc=μ, scale=σ)***
* Calculate required probability
* For discrete estimation (large n): multiply probability by sample size & round
* Plot histogram (if data available) + theoretical density curve
* Shade the region corresponding to the calculated probability
* Display/print results with proper units and interpretation

**Part A: Standard Normal Distribution (Theory Verification)**

**Problem:**

Find P(–1 ≤ Z ≤ 1) for Z ~ N(0,1)
Python Program:

````python
# Experiment 4 - Part A: Standard Normal Probability
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Parameters
mu    = 0
sigma = 1
a     = -1
b     = 1

# Step 2: Calculate probability using CDF
cdf_a = norm.cdf(a, loc=mu, scale=sigma)
cdf_b = norm.cdf(b, loc=mu, scale=sigma)
prob  = cdf_b - cdf_a

# Step 3: Display result
print("=== Standard Normal Distribution ===")
print(f"P({a} ≤ Z ≤ {b})  = {prob:.6f}")
print(f"Percentage       = {prob*100:.3f}%")
print(f"Note: This matches the empirical rule (~68.27%)")

# Step 4: Visualization
x = np.linspace(-3.5, 3.5, 1000)
y = norm.pdf(x, mu, sigma)

plt.figure(figsize=(10, 5))
plt.plot(x, y, 'b-', linewidth=2, label='N(0,1) pdf')
plt.fill_between(x, y, where=(x >= a) & (x <= b), color='skyblue', alpha=0.5,
                 label=f'P({a} ≤ Z ≤ {b}) = {prob:.4f}')

plt.axvline(a, color='red', linestyle='--', alpha=0.7)
plt.axvline(b,  color='red', linestyle='--', alpha=0.7)
plt.axvline(0,  color='black', linestyle='-', alpha=0.4, label='Mean = 0')

plt.title('Standard Normal Distribution – Probability between -1 and 1')
plt.xlabel('Z (standard deviations)')
plt.ylabel('Probability Density')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
````

# **Result and Discussion:**
The code calculates the probability of a standard normal random variable falling between -
1 and 1, which corresponds to the area under the standard normal curve between these
two values.

**Part B: Application – Weekly Wages of 1000 Workmen**

# **Problem:**
Weekly wages ~ N(μ = ₹70, σ = ₹5), n = 1000 workers
Estimate number of workers with wages:
(i) ₹70 to ₹72
(ii) more than ₹75
(iii) less than ₹65
Python Program:

````python
# Experiment 4 - Part B: Weekly Wages Application
from scipy.stats import norm

# Parameters
mu        = 70
sigma     = 5
n_workers = 1000

# (i) Between 70 and 72
p1 = norm.cdf(72, mu, sigma) - norm.cdf(70, mu, sigma)
n1 = round(p1 * n_workers)

# (ii) More than 75
p2 = 1 - norm.cdf(75, mu, sigma)
n2 = round(p2 * n_workers)

# (iii) Less than 65
p3 = norm.cdf(65, mu, sigma)
n3 = round(p3 * n_workers)

print("=== Weekly Wages Analysis (N = 1000) ===")
print(f"(i)  ₹70.00 – ₹72.00   → {n1:4d} workers   ({p1*100:5.2f}%)")
print(f"(ii) Above ₹75.00      → {n2:4d} workers   ({p2*100:5.2f}%)")
print(f"(iii) Below ₹65.00     → {n3:4d} workers   ({p3*100:5.2f}%)")

print("\nInterpretation:")
print("• Symmetric tails (ii & iii) are almost equal → distribution is symmetric")
print("• About 15–16% workers earn unusually low or high wages")
````

**Uniform Distribution – Mean, Variance and PDF Visualization**

# **Aim**

To obtain the mean and variance of a random variable following a uniform distribution over a given interval and to visualize its probability density function (PDF).

# **Objective**

* To derive and compute the theoretical mean and variance of a continuous uniform distribution.
* To plot the PDF of the uniform distribution over the specified interval.
* To mark the location of the mean on the PDF plot for visual interpretation.
* To understand that in a uniform distribution, every value within the interval is equally likely.

# **Theory**
If a continuous random variable X is uniformly distributed over the interval [a, b], then its probability density function (PDF) is:

**f(x) ={1/(b − a) if a ≤ x ≤ b;0, otherwise} **

Mean (Expected Value):

**μ = E[X] = (a + b) / 2**

Variance:

**Var(X) = σ² = (b − a)² / 12**

Standard Deviation:

**σ = (b − a) / √12**

The distribution is rectangular (flat) between a and b, reflecting equal probability for all values in the interval.

# **Algorithm**

* Import required libraries: numpy, matplotlib.pyplot, scipy.stats.uniform
* Define the lower limit (a) and upper limit (b) of the uniform interval
* Compute mean = (a + b)/2
* Compute variance = (b − a)² / 12
* Generate a dense set of x values using np.linspace (slightly extended beyond [a,b] for better visualization)
* Compute the PDF values using uniform.pdf(x, loc=a, scale=b-a)
* Plot the PDF curve
* Fill the region under the curve between a and b to emphasize the support
* Draw a vertical line at the mean
* Add labels, title, legend, and grid
* Display the plot

**Part A: Case Study – Uniform Distribution over [2, 8]**

Problem:

A random variable X is uniformly distributed over the interval [2, 8].
Obtain the mean, variance and plot the PDF curve.

````python
# Exercise 7 - Uniform Distribution [2, 8]
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Step 1: Define interval
a = 2
b = 8

# Step 2: Theoretical mean and variance
mean_uniform = (a + b) / 2
var_uniform  = (b - a)**2 / 12
std_uniform  = np.sqrt(var_uniform)

print("=== Uniform Distribution [2, 8] ===")
print(f"Interval          : [{a}, {b}]")
print(f"Mean (μ)          : {mean_uniform:.2f}")
print(f"Variance (σ²)     : {var_uniform:.2f}")
print(f"Standard Deviation: {std_uniform:.2f}")

# Step 3: Generate x values (slightly wider for visualization)
x = np.linspace(a - 1.5, b + 1.5, 800)

# Step 4: Compute PDF
pdf = uniform.pdf(x, loc=a, scale=b - a)

# Step 5: Plot
plt.figure(figsize=(9, 5.5))
plt.plot(x, pdf, 'royalblue', lw=2.4, label=f'Uniform({a}, {b}) PDF')

# Fill the actual support region
plt.fill_between(x, pdf, where=(x >= a) & (x <= b), color='lightblue', alpha=0.5,
                 label='Equal probability region')

# Mark mean
plt.axvline(mean_uniform, color='red', linestyle='--', linewidth=2.5,
            label=f'Mean = {mean_uniform:.1f}')

plt.title('Probability Density Function – Uniform Distribution [2, 8]')
plt.xlabel('x')
plt.ylabel('f(x)  (Probability Density)')
plt.grid(True, alpha=0.3)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
````

# **Analysis of Time Gaps Using Exponential Distribution**


# **Aim:**  
To calculate probabilities for inter-arrival times and visualize the exponential distribution’s probability density function.

# **Objective:**  

• To calculate the probability that the time between two consecutive requests is less than 5 minutes and greater than 15 minutes.  

• To visualize the exponential distribution by plotting its probability density function (PDF).  

• To shade the areas under the PDF that correspond to the calculated probabilities.

# **Algorithm:**  
* Step 1: Set the mean time (10 minutes) and calculate the rate parameter (λ = 1 / mean time).  
* Step 2: Use the cumulative distribution function (CDF) of the exponential distribution to calculate the probability that the inter-arrival time is less than 5 minutes and greater than 15 minutes.  
* Step 3: Calculate the probability density function (PDF) for the exponential distribution over a specified time range.  
* Step 4: Plot the PDF, and shade the areas corresponding to the calculated probabilities for inter-arrival times less than 5 minutes and greater than 15 minutes.

# **Case:**
Calculate the probability that the time between two consecutive requests, following an exponential distribution with a mean of 10 minutes, is less than 5 minutes and greater than 15 minutes. Plot the probability density function (PDF) and shade the areas corresponding to these probabilities.


````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Parameters
mean_time = 10          # mean time in minutes
lambda_rate = 1 / mean_time   # rate parameter λ = 1 / mean

# Calculate probabilities
prob_less_than_5 = expon.cdf(5, scale=mean_time)          # P(T < 5)
prob_greater_than_15 = 1 - expon.cdf(15, scale=mean_time) # P(T > 15)

# Print probabilities
print(f"Probability that time is less than 5 minutes   : {prob_less_than_5:.4f}")
print(f"Probability that time is greater than 15 minutes: {prob_greater_than_15:.4f}")
# Generate x values and PDF
x = np.linspace(0, 40, 1000)
pdf = expon.pdf(x, scale=mean_time)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the PDF curve
plt.plot(x, pdf, label='Exponential PDF (mean = 10 min)', color='blue', linewidth=2)

# Shade area for P(T < 5)
plt.fill_between(x, pdf, where=(x <= 5), color='skyblue', alpha=0.5,
                 label=f'P(T < 5) = {prob_less_than_5:.4f}')

# Shade area for P(T > 15)
plt.fill_between(x, pdf, where=(x >= 15), color='lightcoral', alpha=0.5,
                 label=f'P(T > 15) = {prob_greater_than_15:.4f}')

# Labels and title
plt.title("Exponential Distribution – Mean Time = 10 minutes")
plt.xlabel("Time between requests (minutes)")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True, alpha=0.3)

# Show plot
plt.show()
````

# **Result and Discussion:**
The probability that the time between two consecutive requests is less than 5 minutes is approximately 0.3935 (39.35%).
The probability that the time exceeds 15 minutes is approximately 0.2231 (22.31%).
These values are consistent with the properties of the exponential distribution — shorter waiting times are more probable, while longer waiting times become exponentially less likely.
The shaded regions under the PDF clearly illustrate these probabilities.

# **Generation and Verification of Wide-Sense Stationary (WSS) Process**

# **Aim**
To generate a sine wave with random phase and verify that it is a Wide-Sense Stationary (WSS) process by analysing its ensemble mean and autocorrelation function.

# **Objectives**

To simulate multiple realizations of the random process X(t) = a·sin(ω₀t + Θ).
To compute the ensemble mean and autocorrelation function from the simulated realizations.
To verify the two conditions of wide-sense stationarity using empirical and theoretical results.


# **Theory**
A stochastic process X(t) is called Wide-Sense Stationary (WSS) if it satisfies the following two conditions:

* **Condition 1** — Constant Mean:
The ensemble mean must be independent of time t:
E[X(t)] = μ (constant for all t)

* **Condition 2** — Lag-Dependent Autocorrelation:
The autocorrelation function must depend only on the time lag τ = t₂ − t₁, not on the absolute time:

R_X(τ) = E[X(t) · X(t + τ)]

For the process X(t) = a·sin(ω₀t + Θ) where Θ ~ U[0, 2π], the theoretical results are derived as follows:

**Mean:**
E[X(t)] = E[a·sin(ω₀t + Θ)] = a · (1/2π) ∫₀²π sin(ω₀t + θ) dθ = 0
Since the integral of a sine over a full period is zero, the mean is 0 for all t. WSS Condition 1 is satisfied.

**Autocorrelation:**
R_X(τ) = E[a·sin(ω₀t + Θ) · a·sin(ω₀(t+τ) + Θ)]
= (a²/2) · cos(ω₀τ)
This result depends only on the lag τ and not on t. WSS Condition 2 is satisfied.
Since both conditions hold, the process X(t) is Wide-Sense Stationary.

**Problem Statement**
Verify whether the random process:
X(t) = a · sin(ω₀t + Θ)
is Wide-Sense Stationary, given:

Amplitude: a = 1
Angular frequency: ω₀ = 2π rad/s (frequency = 1 Hz)
Phase: Θ ~ U[0, 2π] (uniformly distributed random variable)

# **Procedure / Algorith**m

* Define the parameters: amplitude a, angular frequency ω₀, time vector t ∈ [0, 2] s with 500 points, and lag vector τ ∈ [−1, 1] s with 501 points.

Simulate N realizations of X(t) by broadcasting Θ against the time vector to obtain a matrix of shape (500 × N).

* Compute the ensemble mean E[X(t)] by averaging across realizations (axis = 1) at each time instant.

* Fix t₀ = 0 and sweep τ across the lag vector. At each τ, compute the ensemble average of X(t₀) · X(t₀ + τ) to obtain the empirical autocorrelation R_X(τ).
Compute the theoretical autocorrelation R_X(τ) = (a²/2) cos(ω₀τ) for the same lag vector.

````python
import numpy as np
import matplotlib.pyplot as plt

# ================== Parameters ==================
a       = 1.0
omega0  = 2 * np.pi               # 1 Hz sinusoid
N       = 10000                    # number of realizations
t       = np.linspace(0, 2, 500)   # time axis (0 – 2 s)
taus    = np.linspace(-1, 1, 501)  # lag axis

# ================== Realizations ==================
# theta shape: (N,)  →  broadcast with t[:,None] gives X shape: (500, N)
theta = np.random.uniform(0, 2*np.pi, N)
X     = a * np.sin(omega0 * t[:, None] + theta)   # (500, N)

# ================== Ensemble Mean ==================
# Average ACROSS realizations (axis=1) at every time instant
# FIX: original code used axis=0, which averaged across time, not realizations
mean_X = np.mean(X, axis=1)   # shape (500,)  ← should ≈ 0 ∀ t  [WSS check 1]

# ================== Autocorrelation R_X(τ) ==================
# E[X(t₀)·X(t₀+τ)]. We fix t₀=0; result must NOT depend on t₀ for WSS.
autocorr = np.zeros(len(taus))
for i, tau in enumerate(taus):
    Xt0     = a * np.sin(omega0 * 0   + theta)   # X at t₀ = 0
    Xt0_tau = a * np.sin(omega0 * tau  + theta)   # X at t₀ + τ
    autocorr[i] = np.mean(Xt0 * Xt0_tau)          # ensemble average

# Theoretical:  R_X(τ) = (a²/2) cos(ω₀τ)  — valid when Θ ~ U[0,2π]
R_theory = 0.5 * a**2 * np.cos(omega0 * taus)

# ================== Plotting ==================
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# --- Left: ensemble mean across time ---
axes[0].plot(t, mean_X, 'b-', linewidth=2, label='Ensemble mean')
axes[0].axhline(0, color='r', linestyle='--', linewidth=1,
               label='Theoretical mean = 0')   # WSS condition 1
axes[0].set_title('Ensemble Mean  E[X(t)]')
axes[0].set_xlabel('Time t (s)')
axes[0].set_ylabel('Mean')
axes[0].set_ylim(-0.15, 0.15)
axes[0].legend()
axes[0].grid(True, alpha=0.3)
# --- Right: autocorrelation ---
axes[1].plot(taus, autocorr, 'b-', linewidth=2, label='Empirical  R̂ₓ(τ)')
axes[1].plot(taus, R_theory, 'r--', linewidth=2,
           label=r'Theoretical  $\frac{a^2}{2}\cos(\omega_0\tau)$')
axes[1].set_title(r'Autocorrelation  $R_X(\tau)$')
axes[1].set_xlabel(r'Lag $\tau$ (s)')
axes[1].set_ylabel(r'$R_X(\tau)$')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

fig.suptitle(
    r'WSS Verification:  $X(t)=a\sin(2\pi t+\Theta),\;\Theta\sim U[0,2\pi]$'
    '\nCondition 1: Constant Mean  |  Condition 2: $R_X$ depends only on $\\tau$',
    fontsize=13, fontweight='bold'
)
plt.tight_layout()
plt.show()
````

## Modelling and Visualization of Events Using Exponential Inter-Arrival Times

---

### Aim
To simulate a Poisson arrival process by generating inter-arrival times using the exponential distribution and to visualize both the event times and the distribution of inter-arrival times for better understanding of the stochastic behaviour of such processes.

---

### Objectives
- To simulate a Poisson process by generating inter-arrival times using the exponential distribution with a specified rate λ.
- To calculate event times by computing the cumulative sum of the inter-arrival times and visualize them using a step plot.
- To verify the exponential nature of the inter-arrival times by plotting their histogram.

---

### Theory

A **Poisson process** is a stochastic process that models the occurrence of random events over time. It is characterized by:

1. **Arrival Rate λ:** The average number of events per unit time.
2. **Inter-arrival Times:** The time between consecutive events follows an **Exponential distribution** with mean `1/λ`.
3. **Memoryless Property:** The exponential distribution satisfies `P(T > s+t | T > s) = P(T > t)`, meaning past history does not affect future arrivals.

**PDF of inter-arrival time T:**

$$f_T(t) = \lambda e^{-\lambda t}, \quad t \geq 0$$

**Mean:** $E[T] = \frac{1}{\lambda}$ &nbsp;&nbsp;&nbsp; **Variance:** $\text{Var}(T) = \frac{1}{\lambda^2}$

The actual event times are obtained by the **cumulative sum** of inter-arrival times:

$$t_n = \sum_{k=1}^{n} T_k$$

---

### Algorithm

1. Set the arrival rate λ and define the total number of events to simulate.
2. Generate inter-arrival times using the exponential distribution with mean `1/λ`.
3. Calculate the cumulative sum of the inter-arrival times to get the actual event times.
4. Plot the event times to visualize the arrival process.
5. Create a histogram to verify the exponential distribution of inter-arrival times.
6. Overlay the theoretical exponential PDF on the histogram for comparison.

---

## Part 1 — Base Case
**Case:** Generate 1000 inter-arrival times using the exponential distribution with rate λ = 2 customers per unit time.

````python
import numpy as np
import matplotlib.pyplot as plt

# ================== Parameters ==================
lambda_rate = 2       # Arrival rate (customers per unit time)
num_events  = 1000    # Number of arrivals

# ================== Generate Inter-Arrival Times ==================
# Exponential distribution with mean = 1/lambda
inter_arrival_times = np.random.exponential(scale=1/lambda_rate,
                                             size=num_events)

# ================== Compute Event Times ==================
# Cumulative sum gives the actual arrival timestamps
event_times = np.cumsum(inter_arrival_times)

# ================== Plotting ==================
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# --- Left: Step plot of event times ---
axes[0].plot(event_times, np.arange(1, num_events + 1),
             drawstyle='steps-post', color='steelblue', linewidth=1.2)
axes[0].set_title('Event Times of Poisson Process', fontsize=13)
axes[0].set_xlabel('Time')
axes[0].set_ylabel('Event Number (Cumulative Count)')
axes[0].grid(True, alpha=0.3)

# --- Right: Histogram with theoretical PDF overlay ---
axes[1].hist(inter_arrival_times, bins=40, density=True,
             alpha=0.7, color='skyblue', edgecolor='black',
             label='Empirical histogram')

# Overlay theoretical exponential PDF
t_pdf = np.linspace(0, inter_arrival_times.max(), 300)
pdf   = lambda_rate * np.exp(-lambda_rate * t_pdf)
axes[1].plot(t_pdf, pdf, 'r-', linewidth=2,
             label=r'Theoretical PDF: $\lambda e^{-\lambda t}$')

axes[1].set_title('Histogram of Inter-Arrival Times', fontsize=13)
axes[1].set_xlabel('Inter-Arrival Time')
axes[1].set_ylabel('Density')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

fig.suptitle(
    r'Poisson Process Simulation  ($\lambda = 2$, $N = 1000$ events)',
    fontsize=13, fontweight='bold'
)
plt.tight_layout()
plt.show()

# ================== Summary Statistics ==================
print("=" * 50)
print("       BASE CASE SUMMARY  (lambda = 2)")
print("=" * 50)
print(f"  Theoretical mean inter-arrival : {1/lambda_rate:.4f}")
print(f"  Empirical   mean inter-arrival : {np.mean(inter_arrival_times):.4f}")
print(f"  Theoretical std  inter-arrival : {1/lambda_rate:.4f}")
print(f"  Empirical   std  inter-arrival : {np.std(inter_arrival_times):.4f}")
print(f"  Total simulation time          : {event_times[-1]:.2f} units")
print("=" * 50)
````

````python
import numpy as np
import matplotlib.pyplot as plt

# ================== Parameters ==================
lambda_service = 2     # 2 customers per unit time
num_customers  = 1000

# ================== Simulation ==================
inter_arrival_service = np.random.exponential(scale=1/lambda_service,
                                               size=num_customers)
arrival_times_service = np.cumsum(inter_arrival_service)

# ================== Plotting ==================
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# --- Left: Arrival step plot ---
axes[0].plot(arrival_times_service, np.arange(1, num_customers + 1),
             drawstyle='steps-post', color='darkorange', linewidth=1.2)
axes[0].set_title('Customer Arrivals in Service Center', fontsize=13)
axes[0].set_xlabel('Time (units)')
axes[0].set_ylabel('Cumulative Customer Count')
axes[0].grid(True, alpha=0.3)

# --- Right: Histogram + PDF ---
axes[1].hist(inter_arrival_service, bins=40, density=True,
             alpha=0.7, color='orange', edgecolor='black',
             label='Empirical histogram')

t_pdf = np.linspace(0, inter_arrival_service.max(), 300)
pdf   = lambda_service * np.exp(-lambda_service * t_pdf)
axes[1].plot(t_pdf, pdf, 'r-', linewidth=2,
             label=r'Theoretical PDF: $\lambda e^{-\lambda t}$')

axes[1].set_title('Inter-Arrival Time Distribution', fontsize=13)
axes[1].set_xlabel('Inter-Arrival Time (units)')
axes[1].set_ylabel('Density')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

fig.suptitle(
    r'Service Center Arrivals  ($\lambda = 2$ customers/unit time, $N = 1000$)',
    fontsize=13, fontweight='bold'
)
plt.tight_layout()
plt.show()

# ================== Summary ==================
print("=" * 55)
print("   APPLICATION 1 — SERVICE CENTER (lambda = 2)")
print("=" * 55)
print(f"  Theoretical mean inter-arrival : {1/lambda_service:.4f} units")
print(f"  Empirical   mean inter-arrival : {np.mean(inter_arrival_service):.4f} units")
print(f"  Total customers simulated      : {num_customers}")
print(f"  Total simulation time          : {arrival_times_service[-1]:.2f} units")
print("=" * 55)
````

````python
import numpy as np
import matplotlib.pyplot as plt

# ================== Parameters ==================
lambda_hospital = 3    # 3 patients per hour
num_patients    = 1000

# ================== Simulation ==================
inter_arrival_hospital = np.random.exponential(scale=1/lambda_hospital,
                                                size=num_patients)
arrival_times_hospital = np.cumsum(inter_arrival_hospital)

# ================== Plotting ==================
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# --- Left: Arrival step plot ---
axes[0].plot(arrival_times_hospital, np.arange(1, num_patients + 1),
             drawstyle='steps-post', color='mediumseagreen', linewidth=1.2)
axes[0].set_title('Patient Arrivals — Hospital Emergency Dept.', fontsize=13)
axes[0].set_xlabel('Time (hours)')
axes[0].set_ylabel('Cumulative Patient Count')
axes[0].grid(True, alpha=0.3)

# --- Right: Histogram + PDF ---
axes[1].hist(inter_arrival_hospital, bins=40, density=True,
             alpha=0.7, color='mediumseagreen', edgecolor='black',
             label='Empirical histogram')

t_pdf = np.linspace(0, inter_arrival_hospital.max(), 300)
pdf   = lambda_hospital * np.exp(-lambda_hospital * t_pdf)
axes[1].plot(t_pdf, pdf, 'r-', linewidth=2,
             label=r'Theoretical PDF: $\lambda e^{-\lambda t}$')

axes[1].set_title('Inter-Arrival Time Distribution', fontsize=13)
axes[1].set_xlabel('Inter-Arrival Time (hours)')
axes[1].set_ylabel('Density')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

fig.suptitle(
    r'Hospital ED Patient Arrivals  ($\lambda = 3$ patients/hour, $N = 1000$)',
    fontsize=13, fontweight='bold'
)
plt.tight_layout()
plt.show()

# ================== Summary ==================
print("=" * 55)
print("  APPLICATION 2 — HOSPITAL EMERGENCY (lambda = 3)")
print("=" * 55)
print(f"  Theoretical mean inter-arrival : {1/lambda_hospital:.4f} hours")
print(f"  Empirical   mean inter-arrival : {np.mean(inter_arrival_hospital):.4f} hours")
print(f"  Theoretical std  inter-arrival : {1/lambda_hospital:.4f} hours")
print(f"  Empirical   std  inter-arrival : {np.std(inter_arrival_hospital):.4f} hours")
print(f"  Total patients simulated       : {num_patients}")
print(f"  Total simulation time          : {arrival_times_hospital[-1]:.2f} hours")
print(f"  Approx. patients per hour      : {num_patients/arrival_times_hospital[-1]:.2f}")
print("=" * 55)
````
