Here we learn the difference between Mean (Average) and Median (Middle value).

### Finding the Groups (Filtering Data)
Now we use statistics to take action. We use quartiles for Top, middle and lower end groups.

###  Gender Analysis

The Principal has asked a question: "Who performed better on average? The Boys or the Girls?" If we just look at the whole class average, we miss this detail. We need to split the data!

- Filtering: We can create separate lists for Boys and Girls.

- Grouping: We use the groupby commandâ€”the Swiss Army Knife of Pandasâ€”to calculate statistics for each group instantly.

````python
print("\n CHAPTER 8: BOYS vs GIRLS (Group Comparison)")
print("-" * 60)

# METHOD 1: The "Manual" Way (Filtering)
# Let's create a dataframe just for Boys
boys_data = loaded_data[loaded_data['Gender'] == 'M']
girls_data = loaded_data[loaded_data['Gender'] == 'F']

print(f"Number of Boys: {boys_data.shape[0]}")
print(f"Number of Girls: {girls_data.shape[0]}")

# Let's check Math averages manually
boy_math_avg = boys_data['Math'].mean()
girl_math_avg = girls_data['Math'].mean()

print(f"\nManual Check - Math Average:")
print(f"Boys: {boy_math_avg:.2f}")
print(f"Girls: {girl_math_avg:.2f}")

# METHOD 2: The "Pro" Way (Groupby)
# Imagine calculating this for 50 columns! Manual is too slow.
# .groupby() does it all at once.
print("\nThe Power of GroupBy (All Subjects at once):")

# numeric_only=True tells pandas to ignore names/IDs and only calculate math
gender_performance = loaded_data.groupby('Gender').mean(numeric_only=True)
display(gender_performance)

print("\nDETECTIVE INSIGHT:")
print("Look at the Math scores above.")
print("The Boys' average is significantly lower.")
print("Is it because ALL boys are bad at math?")
print("NO! Remember 'Kevin' (15 marks) is a Boy.")
print("One outlier can drag down the whole group's reputation!")

# Let's look at Median to be fair (Median ignores Kevin)
print("\nLet's check the Median (Removing the 'Kevin Effect'):")
gender_median = loaded_data.groupby('Gender').median(numeric_only=True)
display(gender_median)
print("Notice how the gap closes when we use Median?")
````

# Measures of Central Tendency with Visualization

---



## Aim  
To compute the **mean**, **median**, and **mode** of a given dataset using Python, and to visualize the data distribution using **histogram** and **boxplot** to understand central tendency and spread.

## Objective  
1. To calculate the three primary measures of central tendency:  
   - Mean (arithmetic average)  
   - Median (middle value)  
   - Mode (most frequent value)  
2. To implement these calculations manually (without built-in functions where possible) for better conceptual understanding.  
3. To visualize the dataset using histogram (to show frequency distribution) and boxplot (to show median, quartiles, range, and outliers).  
4. To interpret how outliers affect mean vs. median, and how mode reflects repetition in data.  
5. To relate these measures to real-world engineering applications (e.g., quality control, sensor data analysis, reliability studies).

## Theory (Brief)  
- **Mean**: Sum of all observations divided by the number of observations. Sensitive to extreme values (outliers).  
  Formula: Î¼ = (Î£xáµ¢) / n  
- **Median**: The middle value when data is arranged in ascending order. Robust to outliers.  
  - If n is odd â†’ middle term  
  - If n is even â†’ average of two middle terms  
- **Mode**: The value that appears most frequently in the dataset. A dataset can be unimodal, bimodal, or multimodal.  
- **Histogram**: Graphical representation of frequency distribution.  
- **Boxplot**: Summarizes data using five-number summary (minimum, Q1, median, Q3, maximum) and shows outliers.

## Algorithm  
1. Import necessary libraries:  
   - `numpy` for numerical operations  
   - `matplotlib.pyplot` for visualization  
   - `collections.Counter` for frequency count (mode)  
2. Define the given dataset as a list.  
3. Compute **Mean**:  
   - Sum all elements using `np.sum()`  
   - Divide by the length of the list (`len(data)`)  
4. Compute **Median**:  
   - Sort the data in ascending order  
   - If length is odd â†’ take middle element  
   - If length is even â†’ take average of two middle elements  
5. Compute **Mode**:  
   - Use `Counter` to count frequency of each value  
   - Find the value with highest frequency using `most_common(1)`  
6. Display the calculated mean, median, and mode with proper formatting.  
7. Visualize the data:  
   - Plot a **histogram** to show frequency distribution (choose appropriate bins)  
   - Plot a **boxplot** to show central value, spread, and possible outliers  
8. Interpret the results:  
   - Compare mean vs. median (to check effect of outliers)  
   - Note the mode (most frequent value)  
   - Analyze the shape of histogram and boxplot

````python
# ============================================================
# Exercise No: 2
# Measures of Central Tendency with Visualization
# ============================================================

# ------------------------------------------------------------
# GIVEN DATA
data = [12, 15, 18, 20, 22, 15, 18, 25, 30, 18]
print("Given Data:", data)

# ------------------------------------------------------------
# IMPORT REQUIRED LIBRARIES
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# ------------------------------------------------------------
# FUNCTION TO CALCULATE MEAN
def calculate_mean(data):
    """
    Mean = (Sum of all observations) / (Number of observations)
    """
    return np.sum(data) / len(data)

# ------------------------------------------------------------
# FUNCTION TO CALCULATE MEDIAN
def calculate_median(data):
    """
    Median is the middle value of the sorted data
    """
    sorted_data = sorted(data)
    n = len(sorted_data)

    if n % 2 == 0:
        median = (sorted_data[n//2 - 1] + sorted_data[n//2]) / 2
    else:
        median = sorted_data[n//2]

    return median

# ------------------------------------------------------------
# FUNCTION TO CALCULATE MODE
def calculate_mode(data):
    """
    Mode is the value with highest frequency
    """
    frequency = Counter(data)
    mode = frequency.most_common(1)[0][0]
    return mode

# ------------------------------------------------------------
# CALCULATE MEASURES OF CENTRAL TENDENCY
mean_value = calculate_mean(data)
median_value = calculate_median(data)
mode_value = calculate_mode(data)

# ------------------------------------------------------------
# DISPLAY RESULTS
print("\nMeasures of Central Tendency:")
print("Mean   :", mean_value)
print("Median :", median_value)
print("Mode   :", mode_value)

# ------------------------------------------------------------
# VISUALIZATION
# ------------------------------------------------------------

# HISTOGRAM â€“ Data Distribution
plt.figure()
plt.hist(data, bins=6)
plt.xlabel("Data Values")
plt.ylabel("Frequency")
plt.title("Histogram of Given Data")
plt.show()

# BOXPLOT â€“ Spread and Central Value
plt.figure()
plt.boxplot(data, vert=False)
plt.xlabel("Data Values")
plt.title("Boxplot of Given Data")
plt.show()

# ------------------------------------------------------------
# INTERPRETATION (FOR STUDENTS):
# - Histogram shows frequency distribution of data
# - Boxplot shows median, range, and possible outliers
# - Mean is affected by extreme values
# - Median gives the central value
# - Mode shows the most repeated observation

# ------------------------------------------------------------
# CONCLUSION:
# Statistical measures along with visualization provide
# a clear understanding of the data distribution and
# central tendency in engineering applications.
# ============================================================
````

# **Measures of Central Tendency using Pandas & NumPy**
# (From Excel Data â€“ Engineering Mathematics Lab)

````python
# ============================================================
# Exercise No: 2
# Measures of Central Tendency using Pandas & NumPy
# (From Excel Data â€“ Engineering Mathematics Lab)
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ------------------------------------------------------------
# AIM:
# To calculate Mean, Median, and Mode from data read
# from an Excel file and visualize the data.

# ------------------------------------------------------------
# READ DATA FROM EXCEL FILE
filename = "/content/drive/MyDrive/Marks.xlsx"

# Read Excel file (no header assumed)
df = pd.read_excel(filename, header=None)

# Convert DataFrame to NumPy array
data = df.values.astype(float)

# Transpose data (each row represents one dataset)
D = data.T

print("Original Data:\n", data)
print("\nTransposed Data:\n", D)

# ------------------------------------------------------------
# MEASURES OF CENTRAL TENDENCY
# ------------------------------------------------------------

# 1. MEAN
# Mean = Average value of the data
mean_values = np.mean(D, axis=1)

# 2. MEDIAN
# Median = Middle value after sorting
median_values = np.median(D, axis=1)

# 3. MODE
# Mode = Most frequently occurring value
def calculate_mode(arr):
    values, counts = np.unique(arr, return_counts=True)
    return values[np.argmax(counts)]

mode_values = np.apply_along_axis(calculate_mode, 1, D)

# ------------------------------------------------------------
# DISPLAY RESULTS
print("\nMeasures of Central Tendency (Row-wise after Transpose):")
for i in range(len(mean_values)):
    print(f"\nData Set {i+1}:")
    print("Mean   =", mean_values[i])
    print("Median =", median_values[i])
    print("Mode   =", mode_values[i])

# ------------------------------------------------------------
# VISUALIZATION
# ------------------------------------------------------------

# HISTOGRAM â€“ Distribution of marks
plt.figure()
plt.hist(D.flatten(), bins=10)
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Histogram of Marks Distribution")
plt.show()

# BOXPLOT â€“ Spread and outliers
plt.figure()
plt.boxplot(D.flatten(), vert=False)
plt.xlabel("Marks")
plt.title("Boxplot of Marks")
plt.show()


````

# **Theory (Grouped Data)**

When data is given in class intervals with frequencies, individual values are not directly available.
So we use class mid-points.

````python
import math

def get_grouped_data():
    """
    Takes class intervals and frequencies as input
    """
    n = int(input("Enter number of class intervals: "))

    classes = []
    frequencies = []

    for i in range(n):
        print(f"\nClass interval {i+1}")
        lower = float(input("Enter lower limit: "))
        upper = float(input("Enter upper limit: "))
        freq = int(input("Enter frequency: "))

        classes.append((lower, upper))
        frequencies.append(freq)

    return classes, frequencies


def calculate_midpoints(classes):
    midpoints = []
    for c in classes:
        mid = (c[0] + c[1]) / 2
        midpoints.append(mid)
    return midpoints


def calculate_mean(midpoints, frequencies):
    total_fx = 0
    total_f = sum(frequencies)

    for i in range(len(midpoints)):
        total_fx += midpoints[i] * frequencies[i]

    return total_fx / total_f


def calculate_std_dev_grouped(midpoints, frequencies):
    """
    Ïƒ = sqrt( Î£ f(x âˆ’ xÌ„)Â² / Î£ f )
    """
    mean = calculate_mean(midpoints, frequencies)
    total_f = sum(frequencies)

    sum_fxd2 = 0
    for i in range(len(midpoints)):
        sum_fxd2 += frequencies[i] * (midpoints[i] - mean) ** 2

    variance = sum_fxd2 / total_f
    return math.sqrt(variance)


def display_results(classes, frequencies, midpoints, std_dev):
    print("\nClass Intervals :", classes)
    print("Frequencies    :", frequencies)
    print("Midpoints      :", midpoints)
    print("Standard Deviation :", round(std_dev, 2))


# -------- Program Execution --------
classes, frequencies = get_grouped_data()
midpoints = calculate_midpoints(classes)
std_dev = calculate_std_dev_grouped(midpoints, frequencies)
display_results(classes, frequencies, midpoints, std_dev)
````

The measures of dispersion provide insights into the spread and variability within the
given dataset. A higher standard deviation suggests a wider spread of data points, while a
lower standard deviation indicates that data points are clustered more closely around the
mean. In this case, a standard deviation of 4.96 suggests moderate variability within the
dataset.

# Calculate the probability of successes of a random variable which follows Binomial distribution

---



=============================================================================
Experiment: Binomial Probability using scipy.stats
# **Aim:**
  To calculate the probability of obtaining exactly 'x' successes
  in 'n' independent Bernoulli trials using the Binomial distribution.

# **Objective:**

  Given:

- n = number of trials
- p = probability of success in each trial
- x = number of successes we want
  Compute the probability P(X = x) using the binomial probability mass function.

**Formula (for reference):**

  P(X = x) = C(n,x) Ã— p^x Ã— (1-p)^(n-x)
  
  where C(n,x) is the binomial coefficient "n choose x"


# **Algorithm:**
* Step 1: Import the required function (Import binom function) from scipy.stats.
* Step 2: Define the parameters (as given in the case).
* Step 3: Calculate the probability using binom.pmf(), ie, Calculate exact probability P(X = x).
* Step 4: Display the result in a clear, formatted way.
* Step 4: Visualize the complete Probability Mass Function (PMF)

````python
# =============================================================================
# Binomial Probability Distribution
# =============================================================================

# Import required libraries
from scipy.stats import binom
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define parameters
n = 10          # Total number of independent trials
p = 0.5         # Probability of success in each trial (fair coin)
x = 6           # Desired number of successes

# Step 2: Calculate exact probability P(X = x)
probability = binom.pmf(k=x, n=n, p=p)

# Step 3: Display results
print("Binomial Probability Calculation")
print("-" * 50)
print(f"Number of trials (n)          : {n}")
print(f"Probability of success (p)    : {p}")
print(f"Desired number of successes   : {x}")
print("-" * 50)
print(f"Probability of getting exactly {x} successes   : {probability:.6f}")
print(f"Formatted (4 decimal places)                  : {probability:.4f}")
print(f"Scientific notation                           : {probability:.2e}")
print("-" * 50)

# Step 4: Visualize the complete Probability Mass Function (PMF)
# Generate all possible values of x (0 to n)
possible_x = np.arange(0, n + 1)
pmf_values = binom.pmf(possible_x, n, p)

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(possible_x, pmf_values, color='skyblue', edgecolor='black', alpha=0.8)

# Highlight the desired x value
plt.axvline(x=x, color='red', linestyle='--', linewidth=2.5,
            label=f'x = {x} (P = {probability:.4f})')

# Labels and formatting
plt.title(f'Binomial Probability Mass Function\n(n = {n}, p = {p})', fontsize=14, fontweight='bold')
plt.xlabel('Number of Successes (x)', fontsize=12)
plt.ylabel('Probability P(X = x)', fontsize=12)
plt.xticks(possible_x)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(fontsize=11)
plt.tight_layout()

# Show the plot
plt.show()
````

````python
# Alternative 1: Stem + Line plot (very common in probability textbooks)

plt.figure(figsize=(10, 6))
plt.stem(possible_x, pmf_values, linefmt='b-', markerfmt='bo', basefmt=' ',
         label='P(X = x)')
plt.plot(possible_x, pmf_values, 'r--', alpha=0.6, label='Connecting line')

plt.axvline(x=x, color='green', linestyle='-', linewidth=2.5,
            label=f'x = {x} (P = {probability:.4f})')

plt.title(f'Binomial Distribution â€“ Stem & Line Plot\n(n = {n}, p = {p})', fontsize=14)
plt.xlabel('Number of Successes (x)', fontsize=12)
plt.ylabel('Probability P(X = x)', fontsize=12)
plt.xticks(possible_x)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=11)
plt.tight_layout()
plt.show()
````

# **Result and Discussion:**

The result (0.2051) means there's about a 20.51% chance of getting exactly 6 successes (e.g., heads) in 10 trials (e.g., coin flips) when the probability of success in each trial is 0.5 (e.g., a fair coin).

````python
# =============================================================================
# Application Problem 1: Network Packet Transmission
# Binomial Distribution: n = 15 packets, p = 0.95 success probability
# =============================================================================

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n_packets = 15
p_success = 0.95
possible_k = np.arange(0, n_packets + 1)

# Calculate PMF for all possible successes
pmf = binom.pmf(possible_k, n_packets, p_success)

# Required probabilities
p_exactly_13 = binom.pmf(13, n_packets, p_success)
p_less_than_13 = binom.cdf(12, n_packets, p_success)          # P(X â‰¤ 12) = P(X < 13)
p_between_5_and_10 = binom.cdf(10, n_packets, p_success) - binom.cdf(4, n_packets, p_success)

# Display results
print("=== Network Packet Transmission (n=15, p=0.95) ===")
print(f"Mean expected successful packets : {n_packets * p_success:.2f}")
print(f"Variance                          : {n_packets * p_success * (1 - p_success):.4f}")
print("-" * 60)
print(f"1. P(exactly 13 successful)       : {p_exactly_13:.6f}  ({p_exactly_13:.4%})")
print(f"2. P(less than 13 successful)     : {p_less_than_13:.6f}  ({p_less_than_13:.4%})")
print(f"3. P(between 5 and 10 inclusive)  : {p_between_5_and_10:.6f}  ({p_between_5_and_10:.4%})")
print("-" * 60)

# Visualization - Bar plot of PMF
plt.figure(figsize=(12, 6))
bars = plt.bar(possible_k, pmf, color='cornflowerblue', edgecolor='navy', alpha=0.85)

# Highlight interesting regions
for k in [13]:
    bars[k].set_color('darkred')
    bars[k].set_edgecolor('gold')
    bars[k].set_linewidth(2)

plt.title("Binomial PMF â€“ Successful Packet Transmission\n(n = 15, p = 0.95)", fontsize=14, fontweight='bold')
plt.xlabel("Number of Successful Packets (k)", fontsize=12)
plt.ylabel("Probability P(K = k)", fontsize=12)
plt.xticks(possible_k)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.ylim(0, max(pmf)*1.15)

# Add text annotations for key probabilities
plt.text(13, pmf[13]+0.005, f'P(13) = {p_exactly_13:.4f}', ha='center', fontsize=10, color='darkred')

plt.tight_layout()
plt.show()
````
